```python
cities = {
    'karachi': 'Pakistan',
    'chkwal': 'Pakistan',
    'Lahore': 'Pakistan',
    'California': 'USA',
    'Texa'
}

# Initialize an empty dictionary to hold the results
result = {}

# Iterate over the items in the original dictionary
for city, country in cities.items():
    # If the country is not yet a key in the result dictionary, add it with an empty list
    if country not in result:
        result[country] = []
    # Append the city to the list corresponding to its country
    result[country].append(city)

# Print the result dictionary
print(result)
```
