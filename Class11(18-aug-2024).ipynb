```python
def make_album():
    dic = {}
    while True:
        artist_name = input('Enter Artist Name or press "q" to quit: ')
        if artist_name == 'q':
            break
            
        album_name = input('Enter Album Name or press "q" to quit: ')
        if album_name == 'q':
            break
        else:
            dic[album_name] = artist_name
            continue
            
    return dic
```

```python
make_album()
```

# Using Arbitrary Keyword Arguments


```python
def make_album(artist_name, album_title,no_of_songs = None ):
    dic = {}
    dic[artist_name] = album_title
    if no_of_songs:
        dic['Number of Songs'] = no_of_songs
    return dic
```

```python
make_album('imran khan', 'testing')
```

```python
def make_album_new(artist_name, album_title, **other_info ):
    other_info[artist_name] = album_title
    
    return other_info
```

```python
make_album_new('imran khan', 'testing', no_of_songs = 10, genre = 'Hip hop')
```
