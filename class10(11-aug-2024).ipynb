```python
# input
input_dic = {
    'karachi': 'Pakistan', 
    'Chakwal': 'Pakistan',
    'Lahore': 'Pakistan',
    'Istanbul': 'Turkey', 
    'Bursa': 'Turkey', 
    'California': 'USA', 
    'Captown': 'South Africa'
}

# output
{
    'Pakistan': ['Chakwal', 'Lahore'], 
    'USA': ['California', 'Texax', 'Chicago']
}
```

```python
result = {}
for city, country in input_dic.items():
    if country not in result.keys():
        result[country] = []
    result[country].append(city)
        
print(result)
```

# Functions

```python
# Aaam zindagi
num1 = int(input('Enter first number: '))
num2 = int(input("Enter second number: "))

result = num1 + num2
print(result)
```

```python
# Mentos Zindagi
# Created a function
def add(num1, num2): # num1 and num2 are parameters here
    ''' This function will print the sum of two numbers'''
    result = num1+num2
    print(result)
```

```python
# Called a function
add(12,13) # 12 and 13 are arguments here
```

```python
add(32,12)
```

```python
add(12.2, 13.2)
```

```python
add('Ali', 'Khan')
```

# Positional arguments 

```python
 def greet(first_name, middle_name, last_name):
    print(f'{first_name} {middle_name} {last_name}')
```

```python
# Positional arguments
greet('syed', 'tariq', 'hussain')
```

```python
greet('ali', 'khan')
```

```python
# Default value / parameter value
def greet(first_name, last_name, middle_name = '\b'): 
    print(f'{first_name} {middle_name} {last_name}')
```

## T-shirt

```python
def make_shirt(size, text):
    print(f'The size of the shirt is {size} and {text}')
```

```python
# Positional arguments
make_shirt(12, 'The Boys')
```

```python
# keyword argument
make_shirt(text = 'hello world', size = 12)
```

## Album

```python
def make_album(artist_name, album_title):
    dic = {}
    dic[artist_name] = album_title
    return dic
```

```python
make_album('Chahat Fateh', 'Bado Badi')
```

```python
print(make_album('Imran Khan', 'Unforgettable'))
print(make_album('Shubh', 'Still Rollin'))
print(make_album('Prophec', 'Midnight paradise'))
```

```python
def make_album(artist_name, album_title,no_of_songs = None ):
    dic = {}
    dic[artist_name] = album_title
    if no_of_songs:
        dic['Number of Songs'] = no_of_songs
    return dic
```

```python
make_album('Imran Khan', 'Unforgettable')
```

```python
make_album('Imran Khan', 'Unforgettable', 10)
```
