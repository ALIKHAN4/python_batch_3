# Multiple Conditions

```python
name = input('Enter student name: ')
marks = int(input("Enter student's marks: "))
total = int(input("Enter total marks"))

per = (marks/total)*100

if per >= 80:
    print(f'Grade: A+, Percentage: {per}')
elif 70 <= per < 80:
    print(f'Grade: A, Percentage: {per}')
elif 60 <= per < 70:
    print(f'Grade: B, Percentage: {per}')
elif 50 <= per < 60:
    print(f'Grade: C, Percentage: {per}')
elif 40 <= per < 50:
    print(f'Grade: D, Percentage: {per}')
else:
    print('Fail')

    
```

```python
income = int(input('Enter Yearly income: '))

if 600000 < income <= 1200000:
    tax = ((income-600000)*5)/100
    cat = 'B'
    
elif 1200000 < income <= 2200000:
    tax = 30000 + (((income-1200000)*15)/100)
    cat = 'C'
    
elif 2200000 < income <= 3200000:
    tax = 180000 + (((income-2200000)*25)/100)
    cat = 'D'
    
elif 3200000 < income <= 4100000:
    tax = 430000 + (((income-3200000)*30)/100)
    cat = 'E'

elif income > 4100000:
    tax = 700000 + (((income-4100000)*35)/100)
    cat = 'F'

else:
    tax = 0
    cat = 'A'
    
print(f'The tax is {tax} and the person lies in category {cat}')
    
```
