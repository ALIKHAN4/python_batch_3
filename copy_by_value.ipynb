# Copy by value and Reference

## Copy By value

```python
a = 'ali khan'
nums = [1,2,3,4]
```

```python
nums_copy = nums.copy()

#nums
#nums_copy

nums.append(5)

```

```python
nums_copy
```

## Copy By Reference

```python
nums
nums_reference_copy = nums

nums.append(5)
```

```python
nums_reference_copy.append(6)
```

```python
print(nums)
print(nums_reference_copy)
```

```python
print(id(nums))
print(id(nums_reference_copy))
```

```python
nums is nums_reference_copy
```

```python
nums is nums_copy
```

is and == are different things in python

- is (two reference point to the same object or not)
- == (relational operator)

# Object Oriented Programming

```python
class Human:
    name = 'Ali Khan'
    course = 'python'
    roll_no = '113'
    
```

```python
mujtaba = Human
zu_noor = Human
```

```python
class Human:
    def __init__(self, name, course, roll_no):
        self.name = name
        self.course = course
        self.roll_no = roll_no
        
class Janwar:
    def __init__(self, name, course, roll_no):
        self.name = name
        self.course = course
        self.roll_no = roll_no
    
```

```python
mujtaba = Human('ali khan', 'python', '29')


mujtaba = Janwar('mujtaba chandio', 'web dev', '30')
mujtaba = Janwar('mujtaba chandio', 'web dev', '30')

```
